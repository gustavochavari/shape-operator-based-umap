## Complexos simpliciais

Para construir o grafo em alta dimensionalidade, o UMAP se baseia na construção do *Complexo de Čech*, que é uma maneira de representar uma topologia combinatoriamente (usando conjuntos em vez de geometria contínua). Um *complexo* é definido como uma decomposição de um espaço topológico em pedaços menores e topologicamente simples, cujas interseções são complexos menores do mesmo tipo. 

Estes pedaços menores, chamados $k$-*simplexos*, são definidos como fecho convexo de quaisquer $k+1$ pontos linearmente independentes $U=\{u_0,u_1,\dots,u_{k+1}\}$ em $\mathbb{R}^n$, sendo $k=0,1,\dots,n$. Também dizemos que o conjunto $U$ gera o simplexo $\sigma$, e por isso, a dimensão de um $k$-simplexo é $\dim U = k$. Uma face $\tau$ de um simplexo $\sigma$, denotada por $\tau \leq \sigma$, é o fecho convexo de qualquer subconjunto de $U$. 

Um *complexo simplicial* é uma coleção finita $K$ de simplexos $\sigma$ tais que 

$\sigma \in K  \text{ e } \tau \leq \sigma \text {}\Rightarrow \tau \in K$

$\sigma,\tau \in K \Rightarrow \sigma \cap \tau \text{ é vazio, ou é uma face, ou ambos.} $

Um *subcomplexo* de $K$ é um complexo simplicial $L \subseteq K$. Um subcomplexo de particular interesse é o $j$-*skeleton* $K^{(j)}$, que consiste dos subcomplexos de $K$ com dimensão menor ou igual à $j$, isto é, $K^{(j)}=\{\sigma \in K : \dim \sigma \leq j \}$.

Ao pensar em um conjunto de dados como um complexo simplicial, podemos capturar uma representação da topologia e, combinando os simplexos de uma maneira específica, obtemos algumas garantias teóricas sobre quão bem ele representa a topologia subjacente.

### Sistemas de conjuntos convexos

Complexos simpliciais podem emergir como padrões de interseção entre coleções de conjuntos convexos. E quanto menor a dimensão $n$ do espaço ambiente, mais restritivos são os padrões de interseção observados. Um teorema que ilustra este fato é o *Teorema de Helly*: Considere uma coleção $F$ de conjuntos convexos e fechados em $\mathbb{R}^n$. Então quaisquer $n+1$ conjuntos de $F$ possui uma interseção comum não vazia se, e somente se, todos os conjuntos de $F$ possuem interseção comum não vazia. A ida deste teorema é o mais interessante, pois se encontrar $n+1$ coleções de subconjuntos convexos fechados em $F \subset \mathbb{R}^n$, então é valido que $\bigcap F_i \neq \emptyset$ para todo $i$.

O conceito de complexo simplicial pode ser abstraído para espaços topológicos mais gerais. Define-se um *complexo simplicial abstrato* $A$ como uma coleção de conjuntos $F$ tais que, se $\alpha \in A$ e $\beta \subseteq A$, então $\beta \in A$. Ou seja, os conjuntos em $A$ são os simplexos abstratos. A dimensão de um simplexo abstrato $\alpha$ é dada pela cardinalidade dos vértices do simplexo, isto é, $\dim \alpha = \# \alpha -1$.

Dada uma coleção de subconjuntos $F$, um complexo simplicial abstrato que podemos formar é o *Nervo* de $F$
$$\text{Nrv} F = \{X \subseteq F : \bigcap X \neq \emptyset\}.$$
Essencialmente, o nervo captura as interseções não vazias entre os elementos da coleção $F$. Geralmente, a coleção $F$ é pensada como uma cobertura de um espaço topológico prévio $X$. Assim, o chamado *Teorema do Nervo* garante que se cada subcomplexo de $F$ for fechado e convexo, então $\text{Nrv} F$ e o espaço topológico $X$ são homotópicamente equivalentes. As condições do teorema podem ser relaxadas sem sacrificar a conclusão, sendo necessário apenas que $\bigcap F$ seja triangularizável, todos os subcomplexos em $F$ sejam fechados e toda interseção comum não vazia de subcomplexos em $F$ seja contratível.

A noção de homotopia pode ser intuitivamente entendida como uma deformação contínua de um espaço topológico em outro, mesmo que ambos não sejam homeomorfos. Ou seja, a homotopia é uma condição mais fraca que a continuidade de espaços topológicos. Pense no caso de um círculo e um cilindro infinito em $\mathbb{R}^3$, que são homotópicos mas não homeomorfos, pois podemos projetar o cilindro no círculo ou esticar o círculo no cilindro. 

De fato, podemos mapear $ \mathbb{S}^1 $ na "base" $z=0$ do cilindro com $f(e^{i\theta}) = (e^{i\theta}, 0)$. A projeção do cilindro no círculo esquece a coordenada $\mathbb{R}$, mantendo apenas a angular $g(e^{i\theta}, z) = e^{i\theta}$.




### Complexos de Cech

São complexos simpliciais cujos subconjuntos convexos da coleção $F$ são bolas em um espaço euclidiano de mesmo raio $r$. Considere um conjunto finito de pontos $S$ em $\mathbb{R}^d$, e escreva $B_x(r)=x+r\mathbb{B}^d$ para a bola fechada de centro $x$ e raio $r$. O *Complexo de Cech* para $S$ e $r$ é o nervo da coleção de bolas, isto é, 
$$\text{Cech}(r)=\{\sigma \subseteq S : \bigcap_{x \in \sigma}B_x(r)\neq\emptyset \}.$$

Este tipo de complexo não necessariamente possui uma realização geométrica em $\mathbb{R}^n$, mas sua estrutura abstrata já é suficiente para trabalhar. Por exemplo, conseguimos identificar os subcomplexos do complexo de Cech com o algoritmo *MiniBall*. Por consequência do Teorema de Helly, quaisquer $n+1$ pontos em $S$ estão contidos em uma bola $B_r$ qualquer se, e somente se, todos os pontos de $S$ estão em $B_r$. Logo, decidir se um subconjunto $\sigma \subseteq S$ pertence ao complexo de Cech, é verificar se $\sigma$ cabe dentro de uma bola qualquer de raio $r$.

Considere a MiniBall como sendo a única menor bola que contém $\sigma$. O raio $r_0$ da miniball satisfaz $$r_0 \leq r \iff \sigma \in \text{Cech}(r).$$ Logo, ao encontrar $r_0$, conseguimos saber se $\sigma \in \text{Cech}(r)$. Observe que a miniball pode ser determinada a partir de uma coleção de $k+1 \leq d+1$ pontos que estão na borda. Esta é a estratégia do algoritmo, ele receber dois subconjuntos $\tau, \eta$ disjuntos de $\sigma$ e retorna a miniball que contém todos os pontos de $\tau$ no seu interior e todos os pontos de $\eta$ em sua fronteira.

### Exemplo algoritmo Miniball (Estratégia Move to Front de Wezl)

```python
import numpy as np
from itertools import combinations
import matplotlib.pyplot as plt
```

```python
def pivot_mb(points, tol=1e-12):
    """
    Implementação do algoritmo pivot_mb para computar a miniball em grandes conjuntos de dados.
    
    Args:
        points: numpy array de shape (n, d)
        tol: tolerância numérica
        
    Returns:
        (center, radius): centro e raio da miniball
    """
    points = np.asarray(points)
    n, d = points.shape
    
    # Casos triviais
    if n == 0:
        return None, 0.0
    elif n == 1:
        return points[0], 0.0
    
    # Embaralha os pontos para melhor desempenho
    L = points.copy()
    np.random.shuffle(L)
    
    # Passo inicial: computa a miniball para um pequeno subconjunto
    t = min(d + 1, n)  # Tamanho inicial do subconjunto
    mb, s = mtf_mb(L[:t], [], tol)
    
    # Fase de pivoteamento
    while True:
        # Encontra o ponto com maior excesso (mais promissor)
        excess, k = max_excess(L, mb, t)
        
        if excess <= tol:
            break  # Todos os pontos estão dentro da bola
            
        # Atualiza a miniball incluindo o ponto pivô
        mb_new, s_new = mtf_mb(L[:s], [L[k]], tol)
        
        # Move o ponto pivô para a frente
        L[[t, k]] = L[[k, t]]
        
        # Atualiza os índices
        t = s + 1
        s = s_new + 1
        mb = mb_new
    
    return mb

def max_excess(points, mb, start_idx):
    """
    Encontra o ponto com maior excesso em relação à bola atual.
    
    Args:
        points: array de pontos
        mb: tupla (center, radius) representando a bola atual
        start_idx: índice a partir do qual procurar
        
    Returns:
        (max_excess, index): maior excesso e índice do ponto correspondente
    """
    if mb is None:
        return float('inf'), 0
    
    center, radius = mb
    max_e = -1
    max_idx = start_idx
    
    for i in range(start_idx, len(points)):
        diff = points[i] - center
        e = np.dot(diff, diff) - radius**2
        if e > max_e:
            max_e = e
            max_idx = i
    
    return max_e, max_idx

def mtf_mb(points, boundary, tol=1e-12):
    """
    Implementação do algoritmo move-to-front para miniball com pontos de fronteira.
    
    Args:
        points: pontos a serem englobados
        boundary: pontos que devem estar no boundary
        tol: tolerância numérica
        
    Returns:
        (center, radius): miniball
        s: tamanho do suporte
    """
    if len(points) == 0:
        return trivial_miniball(boundary), len(boundary)
    
    # Embaralha os pontos para melhor desempenho esperado
    L = points.copy()
    np.random.shuffle(L)
    
    # Inicializa com a miniball dos pontos de boundary
    mb = trivial_miniball(boundary)
    if mb[0] is None:
        mb = (L[0], 0.0)
        support = [L[0]]
    else:
        support = boundary.copy()
    
    s = len(support)
    _ , d = points.shape
    
    for i in range(len(L)):
        point = L[i]
        if not is_inside(point, mb, tol):
            support.append(point)
            mb = trivial_miniball(support)
            s = len(support)
            if s > d + 1:
                break  # Não pode ter mais que d+1 pontos de suporte
    
    return mb, s

def trivial_miniball(points):
    """
    Computa a miniball exata para até d+1 pontos.
    """
    points = np.asarray(points)
    if len(points) == 0:
        return None, 0.0
    elif len(points) == 1:
        return points[0], 0.0
    elif len(points) == 2:
        return circle_from_2_points(points[0], points[1])
    else:
        return circle_from_3_points(points)

def circle_from_2_points(a, b):
    """Bola definida por 2 pontos (diâmetro)"""
    center = (a + b) / 2
    radius = np.linalg.norm(a - b) / 2
    return center, radius

def circle_from_3_points(points):
    """Bola definida por 3 pontos (circunferência circunscrita)"""
    A = points[1:] - points[0]
    b = np.array([np.dot(p, p) - np.dot(points[0], points[0]) for p in points[1:]]) / 2
    try:
        center = np.linalg.solve(A, b)
    except np.linalg.LinAlgError:
        # Pontos colineares, retorna a maior bola entre os pares
        balls = [circle_from_2_points(points[i], points[j]) 
                for i, j in combinations(range(len(points)), 2)]
        return max(balls, key=lambda x: x[1])
    radius = np.linalg.norm(center - points[0])
    return center, radius

def is_inside(point, mb, tol):
    """Verifica se um ponto está dentro da bola"""
    if mb[0] is None:
        return False
    return np.linalg.norm(point - mb[0]) <= mb[1] * (1 + tol)
```

## Vietoris-Rips Complex

No complexo de Cech, testar se quaisquer $d+1$ formam um $d$-simplexo é trabalhoso, pois precisaríamos checar todas as subcoleções. Por isso, seria mais rápido checar apenas subcoleções de 2 elementos e ir adicionando simplexos de maior dimensão quando possível. Esta simplificação dá origem ao complexo de *Vietoris-Rips*:
$$\text{Vietoris-Rips}(r)=\{\sigma \in S : \text{diam} (\sigma) \leq 2r \}$$

```python
import gudhi
import matplotlib.pyplot as plt
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # For 3D plots if needed
```

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n
        self.components = {i: {i} for i in range(n)}  # Initialize each point in its own component


    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]


    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        
        if root_x != root_y:            
            # Perform union
            if self.rank[root_x] > self.rank[root_y]:
                self.parent[root_y] = root_x
                self.components[root_x].update(self.components[root_y])  # Merge the components
                del self.components[root_y]
            else:
                self.parent[root_x] = root_y
                self.components[root_y].update(self.components[root_x])  # Merge the components
                del self.components[root_x]
                if self.rank[root_x] == self.rank[root_y]:
                    self.rank[root_y] += 1
            return root_x, root_y
        return None, None

def construct_calculate(data_array, max_dimension, max_edge_length):
    """
    Constructs a Vietoris-Rips complex, computes persistence, and visualizes the persistence diagram.
    
    Parameters:
    - data_array: numpy array of data points.
    - max_dimension: Maximum dimension of the simplices.
    - max_edge_length: Maximum edge length for the Rips complex.
    """
    try:
        # Check if the input data array is empty; if so, raise an error
        if data_array.size == 0:
            raise ValueError("Input data array is empty.")
        
        # Construct a Vietoris-Rips complex from the data points with the specified maximum edge length
        rips_complex = gudhi.RipsComplex(points=data_array, max_edge_length=max_edge_length)
        print("rips constructed")
        
        # Create a simplex tree from the Rips complex with the specified maximum dimension
        simplex_tree = rips_complex.create_simplex_tree(max_dimension=max_dimension)
        print("simplex constructed")

        # Compute the persistence of the simplex tree (the birth and death of homological features)
        persistence = simplex_tree.persistence()
        print("persistence computed")
        
        # Print the persistence intervals to the console for debugging and verification
        #print(f"Persistence intervals: {persistence}")
                
        # To get the points that contribute to the persistence we need the following.

        n_points = len(data_array)
        uf = UnionFind(n_points)


        # Get an appropriate max filtration value
        max_filtration = max([simplex_tree.filtration(s) for s, _ in simplex_tree.persistence_pairs()], default=0) + 1.0

        output_csv_contributing_points_file = "output.csv" 

        with open(output_csv_contributing_points_file, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["Dimension", "Filtration (Birth)", "Filtration (Death)", "Simplex", "Merged Components"])


            for birth_death, simplex in simplex_tree.persistence_pairs():


                death_simplex = simplex  # The simplex that appears at this stage
                death_filtration = simplex_tree.filtration(death_simplex)  # Extract the correct filtration value


                if len(birth_death) > 1:
                    birth_filtration = simplex_tree.filtration(birth_death)
                else:
                    birth_filtration = 0  # Handles unbounded features


                simplex_indices = tuple(sorted(simplex))


                if len(simplex_indices) == 1:  # 0D feature (single point)
                    writer.writerow([0, birth_filtration, death_filtration, simplex_indices, "N/A"])
                else:  # 1D or higher feature
                    if len(simplex_indices) == 2:  # Edge causing component merge
                        p1, p2 = simplex_indices
                        old_comp, new_comp = uf.union(p1, p2)
                        if old_comp is not None:
                            merged_component = f"Component {uf.components.get(new_comp, old_comp)} merged with {uf.components.get(old_comp, new_comp)}"
                            writer.writerow([len(simplex_indices)-2, birth_filtration, death_filtration, simplex_indices, merged_component])


        print(f"All bar information saved to {output_csv_contributing_points_file}")


        # Return the persistence intervals for further use if needed
        return persistence, simplex

    except Exception as e:
        # Catch any exception that occurs, print an error message, and re-raise the exception
        print(f"An error occurred: {e}")
        raise

def plot_rips_complex(data_array, max_edge_length, title="Vietoris-Rips Complex"):
    """
    Plots the Vietoris-Rips complex at a given max_edge_length.
    
    Parameters:
    - data_array: numpy array of data points.
    - max_edge_length: Maximum edge length for the Rips complex.
    - title: Title of the plot.
    """
    # Create the Rips complex
    rips_complex = gudhi.RipsComplex(points=data_array, max_edge_length=max_edge_length)
    simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)  # Max dimension 2 for edges/triangles

    # Extract edges and triangles
    edges = []
    triangles = []
    for simplex, _ in simplex_tree.get_skeleton(2):
        if len(simplex) == 2:  # Edge
            edges.append(simplex)
        elif len(simplex) == 3:  # Triangle
            triangles.append(simplex)

    # Plotting
    fig = plt.figure(figsize=(10, 8))
    
    # Check if the data is 2D or 3D
    if data_array.shape[1] == 2:
        ax = fig.add_subplot(111)
        # Plot edges
        for edge in edges:
            ax.plot(data_array[edge, 0], data_array[edge, 1], 'r-')
        
        # Plot points
        ax.scatter(data_array[:, 0], data_array[:, 1], color='blue', s=5, label='Points', zorder=4)
        # Plot triangles (if any)
        #for triangle in triangles:
        #    ax.fill(data_array[triangle, 0], data_array[triangle, 1], 'g-', alpha=0.2)
    
    elif data_array.shape[1] == 3:
        ax = fig.add_subplot(111, projection='3d')
        # Plot points
        ax.scatter(data_array[:, 0], data_array[:, 1], data_array[:, 2], color='red', s=2, label='Points')
        # Plot edges
        for edge in edges:
            ax.plot(data_array[edge, 0], data_array[edge, 1], data_array[edge, 2], 'r-', alpha=0.5)
        # Plot triangles (if any)
        for triangle in triangles:
            ax.plot_trisurf(data_array[triangle, 0], data_array[triangle, 1], data_array[triangle, 2], color='green', alpha=0.2)
    else:
        raise ValueError("Data must be 2D or 3D for visualization.")


    ax.set_title(title)
    ax.set_xlim(np.min(data_array[:, 0]) - 1, np.max(data_array[:, 0]) + 1)
    ax.set_ylim(np.min(data_array[:, 1]) - 1, np.max(data_array[:, 1]) + 1)
    ax.set_aspect('equal')
    ax.legend()
    plt.show()


# Estudar o que significa esses gráficos
# Persistent Homology
# Parece que tem relação com dimensão, buracos, e etc.

def persistence_barcodes(persistence, output_file):    
        # Visualize persistence barcode
        f, ax = plt.subplots(figsize=(10, 10))
        #plot the persistence barcodes
        gudhi.plot_persistence_barcode(persistence, axes=ax, legend=True)
        plt.savefig(output_file)
        print(f"Persistence barcode saved as {output_file}")     



def persistence_graph(persistence, output_file):
        # Create a new figure and axis for plotting the persistence diagram with a specified size
        f, ax = plt.subplots(figsize=(10, 10))
        
        # Plot the persistence diagram using Gudhi's built-in function, with a legend
        gudhi.plot_persistence_diagram(persistence, axes=ax, legend=True)
        
        # Save the plot as an image file with the specified output file name
        plt.savefig(output_file)
        
        # Print a confirmation message that the persistence diagram was saved
        print(f"Persistence diagram saved as {output_file}")
```

O complexo de Rips de fato possui essa propriedade. Ou seja, o complexo de Vietoris Rips inclui um n-simplexo quando possui todos os (n-1)-simplexos, por construção.

## Relação com o UMAP

No algoritmo UMAP, cada ponto de dados é considerado como uma amostra de uma variedade de alta dimensão. Podemos pensar em cada ponto como um 0-simplex. Ao estender a partir de cada ponto um raio r e conectar pontos que se sobrepõem, podemos construir complexos de 1, 2, ou mais dimensões. Este complexo simplicial faz um trabalho razoável de aproximação da topologia fundamental do conjunto de dados, explicada pelo *Teorema de Nervo*. No UMAP a maior parte da representação da topologia está, na verdade, sendo realizada pelos simplexos de dimensão 0 e 1. Ao considerar apenas os simplexos 0 e 1, efetivamente construímos um grafo, que pode ser facilmente projetado em outro grafo de menor dimensão.

Os simplexos de Cech e Vietoris Rips necessitam de um raio fixado para cada ponto de dados. Porém, dados de alta dimensão do mundo real apresentam um desafio: escolher o raio do tamanho correto para representar bem a topologia. Um raio muito pequeno e tenderemos a aglomerados locais e pontos isolados. Raios muito grandes criam relações que podem introduzir viés. Esse problema é exacerbado pela maldição da dimensionalidade, onde as distâncias entre pontos se tornam cada vez mais semelhantes em dimensões mais altas. O UMAP resolve esse problema de forma inteligente: em vez de usar um raio fixo, o UMAP usa um raio variável determinado para cada ponto com base na distância até seu k-ésimo vizinho mais próximo. Dentro desse raio local, a conectividade é então "fuzzy" (difusa), tornando cada aresta em uma probabilidade. Como não queremos que nenhum ponto fique completamente isolado, adiciona-se uma restrição de que todos os pontos devem estar conectados pelo menos ao seu ponto vizinho mais próximo. A saída final desse processo é um grafo ponderado não-direcionado, com pesos representando a probabilidade de dois pontos estarem "conectados" em nossa variedade de alta dimensão.

Observe que, com isso, a noção local de distância de cada ponto pode variar completamente da noção de distância experienciada pelos seus vizinhos. Logo, é necessário determinar se dois pontos estão conectados com base em pesos de arestas direcionadas potencialmente diferentes. A UMAP simetriza esta inconsistência calculando a probabilidade de que pelo menos uma das arestas exista.
